# IT Educational Institute

```python
# Files data paths
DF_PATH = "./IT_Educational_Institute/data/processed/1_preprocessed_df.pkl"
ROLES_PATH = "./IT_Educational_Institute/data/raw/survey_results_schema.csv"

# Report figures path 
FIG_DIR = "./IT_Educational_Institute/reports/figures"

# Parameters
NA_STRING = "Not Specified"
TRANSPARENT_STRING = 'rgba(0,0,0,0)'

# Specific columns from Data
ROLE_COLS = ['DevType']
TECH_COLS = ['LanguageHaveWorkedWith', 'DatabaseHaveWorkedWith',
            'PlatformHaveWorkedWith', 'WebframeHaveWorkedWith',
            'MiscTechHaveWorkedWith', 'ToolsTechHaveWorkedWith',
            'NEWCollabToolsHaveWorkedWith']
```

```python
# Loading libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import logging
```

## Read Data and Preprocess

```python
# Reading data
raw_df = pd.read_pickle(DF_PATH)
roles_names = pd.read_csv(ROLES_PATH)
```

```python
raw_df.head()
```

### One-Hot Encoding

```python
# Creating a copy of the DataFrame
df = raw_df.copy()

# Importing MultiLabelBinarizer from sklearn
from sklearn.preprocessing import MultiLabelBinarizer

# Creating an encoded DataFrame by applying One-Hot Encoding
encoded_dfs = {}

for col in ROLE_COLS + TECH_COLS:
    binarizer = MultiLabelBinarizer()
    encoded_df = pd.DataFrame(binarizer.fit_transform(df[col]),
                             columns=binarizer.classes_,
                             index=df[col].index)
    encoded_dfs[col] = encoded_df
```

```python
# Concatenating One-Hot Encoded DataFrame
df = pd.concat(encoded_dfs, axis=1)

# Displaying encoded DataFrame
df
```

### Display Jobs Frequency

```python
# Finding the jobs frequency
jobs_freq = df['DevType'].copy().sum().sort_values(ascending=False).reset_index()

# Displaying jobs frequency
jobs_freq
```

### Display Skills Frequency

```python
# Calculating the sum for all columns except DevType
skills_freq = df.copy().drop('DevType', axis=1).sum().sort_values(ascending=False).reset_index()

# Columns for the skills_freq
skills_freq.columns = ['Group', 'Skills', 'Count']

# Displaying the skills frequency
skills_freq
```

```python
skills_freq[:10]
```

```python
# Importing plotly express
import plotly.express as px

# Creating a treemap of skills freq
fig = px.treemap(skills_freq,
                path = ['Group','Skills'],
                color_continuous_scale='deep',
                values='Count',
                color='Count', 
                title='Skills Frequency TreeMap')
fig.update_layout(width=1400, height=700)
fig.update
fig.show()
fig.write_html(os.path.join(FIG_DIR, 'treemap.html'))
```

### Jobs & Skills Relation

```python
# Creating sorted lists of roles and skills
sorted_roles = df['DevType'].sum().sort_values().index.tolist()
sorted_skills = df.drop('DevType', axis=1).sum().sort_values(ascending=False).droplevel(level=0).index.tolist()
```

```python
# Creating an empty skills variable
skills = []

# For each role, calculate the percentage of each skill
for role in sorted_roles:
    # filter for the role
    role_mask = (df[('DevType', role)] == 1)
    
    # For each column, calculate mean of one-hot encoded --> percentage
    skills_role = pd.concat({tech_col: df.loc[role_mask, tech_col].mean() * 100
                            for tech_col in TECH_COLS})
    
    # Append to original list
    skills.append(skills_role)
```

```python
# Concat and format
skills = pd.concat(skills, axis=1)
skills.columns = sorted_roles
skills = skills.reset_index(level=0, drop=True)
skills = skills.loc[sorted_skills]
skills = skills.T
```

```python
# Displaying the DataFrame 
skills
```

```python
# Importing graph_objects from plotly
import plotly.graph_objects as go

# Plotting a Heatmap of Jobs & Skills Relation
fig = go.Figure(data=go.Heatmap(z=skills, x=skills.columns, y=skills.index, colorscale='magma', ygap=1))
fig.update_layout(width=2500, height=700)
fig.show()
fig.write_html(os.path.join(FIG_DIR, 'heatmap.html'))
```

### Jobs Dendrogram

```python
# Importing figure_factory from plotly
import plotly.figure_factory as ff

# Plotting a Dendrogram
fig = ff.create_dendrogram(skills, labels=skills.index, orientation='left', color_threshold=0)
fig.update_layout(width=900, height=700, showlegend=False)
fig.show()
fig.write_html(os.path.join(FIG_DIR, 'Jobs Dendrogram.html'))
```

The clustering using dendrogram resulted in mostly reasonable groupings except for the "Back-End Developer" which is grouped with "Senior Executive (C-Suite, VP, etc) and "Engineering Manager". I'd like to continue digging deeper into that anomaly. 

### Features Standardization

```python
# importing StandardScaler from sklearn
from sklearn.preprocessing import StandardScaler

# Applying the scaler to the skills DataFrame
std_skills = StandardScaler().fit_transform(skills)

# Creating a DataFrame of standardized skills
std_skills = pd.DataFrame(std_skills, columns=skills.columns, index=skills.index)
```

```python
# Plotting a Heatmap of standardized data
fig = go.Figure(data=go.Heatmap(z=std_skills, x=skills.columns, y=skills.index, colorscale='magma', ygap=1))
fig.update_layout(width=2500, height=800)
fig.show()
fig.write_html(os.path.join(FIG_DIR, 'normalized_heatmap.html'))
```
